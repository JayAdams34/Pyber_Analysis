### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head(20)
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
count_rides = pyber_data_df.groupby("city").count()['ride_id']
```

```python
# 2. Get the total drivers for each city type
sum_drivers = pyber_data_df.groupby("city").sum()["driver_count"]
```

```python
#  3. Get the total amount of fares for each city type
sum_fare = pyber_data_df.groupby("city").sum()["fare"]
sum_fare
```

```python
#  4. Get the average fare per ride for each city type. 
sum_fare/count_rides
```

```python
# 5. Get the average fare per driver for each city type. 
sum_fare/sum_drivers
```

```python
#  6. Create a PyBer summary DataFrame.
count_rides = pyber_data_df.groupby("type").count()['ride_id']
sum_drivers = pyber_data_df.groupby("type").sum()["driver_count"]
sum_fare = pyber_data_df.groupby("type").sum()["fare"]
avg_per_ride = sum_fare/count_rides
avg_per_driver = sum_fare/sum_drivers

pyber_data_summary_df = pd.DataFrame({
    "Total Rides": count_rides,
    "Total Drivers": sum_drivers,
    "Total Fares": sum_fare,
    "Average Fare per Ride": avg_per_ride,
    "Average Fare per Driver": avg_per_driver})
pyber_data_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_data_summary_df.index.name = None
pyber_data_summary_df
```

```python
#  8. Format the columns.
pyber_data_summary_df["Total Rides"] = pyber_data_summary_df["Total Rides"].map("{:,}".format)
pyber_data_summary_df["Total Drivers"] = pyber_data_summary_df["Total Drivers"].map("{:,}".format)
pyber_data_summary_df["Total Fares"] = pyber_data_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_data_summary_df["Average Fare per Ride"] = pyber_data_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_data_summary_df["Average Fare per Driver"] = pyber_data_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
type_date_fare = pyber_data_df.groupby(["type", "date"]).sum()["fare"]
type_date_fare
```

```python
# 3. Reset the index on the DataFrame you created in #2. This is needed to use the 'pivot()' function.
type_date_fare = type_date_fare.reset_index()
type_date_fare
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
tdf_pivot = type_date_fare.pivot(index='date', columns='type', values='fare')
```

```python
tdf_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
tdf_loc = tdf_pivot.loc[(tdf_pivot.index >= "2019-01-01") & (tdf_pivot.index <= "2019-04-29")]
tdf_loc
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
tdf_loc.index = pd.to_datetime(tdf_loc.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
tdf_loc.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
tdf_resample = tdf_loc.resample("W").sum()
tdf_resample
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
fig, ax = plt.subplots(figsize=(15,15))
ax.plot(tdf_resample.index, tdf_resample['Rural'], label='Rural')
ax.plot(tdf_resample.index, tdf_resample['Suburban'], label='Subrban')
ax.plot(tdf_resample.index, tdf_resample['Urban'], label='Urban')
ax.legend()
ax.set_xlabel('Dates')
ax.set_ylabel('Fare ($USD)')
ax.set_title("Total Fare by City Type")
# tdf_resample.plot()
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
plt.savefig('analysis/PyBer_fare_summary.png')

```
